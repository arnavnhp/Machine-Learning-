## Inner Product

The inner product is the scalar multiplication of one vector (or matrix) and the transpose of another vector (or matrix). If both arrays are 1D arrays, their dimensions should be identical. If either or both arrays are higher-dimensional, then the last dimensions of both arrays should be identical.

```python
a = np.array([[1,2,3], [4,5,6], [8,9,4]]) # 3 X 3
b = np.array([7,8,9]) # 1 X 1
c = np.inner(a,b)
print(c)  # 1 X 3
```

```python
a.dot(b.T)
```

## Outer Product

Outer product is the dot product of a column vector of size Mx1 and a row vector of size 1xN. The resulting array is a matrix of size MxN.

```python
a = np.array([1,2,3,5,6]) # 5 x 1
b = np.array([5,9,11])  # 1 x 3

c = np.outer(a,b)
print(c)  # 5 x 3
```

## Matrix Power

* a general method to obtain either positive or negative powers of a given square matrix

* first negative power of matrix is technically termed its inverse

* can be used to find inverse or any power of matrix




```python
a = np.random.random([4,4])
a_2 = np.linalg.matrix_power(a,2)
a_inv_2 = np.linalg.matrix_power(a,-2)

print("matrix of pow 2\n", a_2)
print("matrix square inverse\n", a_inv_2)
```

## Singular value Decomposition

one of great dimension reduction algorithms, defines principal components and arranges them by rank

```python
from skimage import data
image = data.astronaut()
print(image.dtype, image.min(), image.max(), image.shape)
plt.imshow(image)
plt.show()
```

```python
# normalize the image by dividing it by maximum value ( here it is 255 )
# reorder the axes to be (3,400,600)

img = image/255.0
img = np.transpose(img, axes=(2,0,1))
```

```python
U,S,V = np.linalg.svd(img)
print(U.shape, S.shape, V.shape)
# first dimension 3 refers to three color channels
```

```python
# S matrix should have dimensions for matrix multiplication
sigma = np.zeros((3,512,512))
for i in range(3):
  np.fill_diagonal(sigma[i,:,:], S[i,:])
print(sigma.shape)
```

```python
# reconstruct original image without dimension reduction
reconstruct = U @ sigma @ V
reconstruct = np.transpose(reconstruct,axes=(1,2,0))
plt.imshow(reconstruct)
plt.show()
```

```python
k = 25
reconstruct = U @ sigma[:,:,:k] @ V[:,:k,:]
reconstruct = np.transpose(reconstruct,axes=(1,2,0))

plt.imshow(reconstruct)

plt.show()
```
